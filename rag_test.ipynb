```python
full_text = "\n\n".join(
    f"[PAGE {doc.metadata.get('page', i+1)}]\n{doc.page_content}"
    for i, doc in enumerate(docs)
)
```

```python
full_text
```

```python
import re

heading_pattern = re.compile(
    re.compile(r"(?m)^(\d+\.\d+)\s+(.+)$")
)


matches = list(heading_pattern.findall(full_text))

for i, match in enumerate(matches):
    print(match)
```

```python
from langchain_core.documents import Document
parent_docs = []

for i, match in enumerate(matches):
    start = match.start()
    end = matches[i + 1].start() if i + 1 < len(matches) else len(full_text)

    heading = match.group(0).strip()
    section_text = full_text[start:end]

    parent_docs.append(
        Document(
            page_content=section_text,
            metadata={
                "parent_id": f"parent_{i}",
                "heading": heading,
                "doc_id": "warehouse_sop",
            }
        )
    )


```

```python
filtered_parents = []
for doc in parent_docs:
    text = doc.page_content.lower()

    if "table of contents" in text[:500]:
        continue
    filtered_parents.append(doc)


```

```python
parent_docs = filtered_parents
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

child_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100
)

child_docs = child_splitter.split_documents(parent_docs)
```

```python
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
load_dotenv()

embeddings = OpenAIEmbeddings(model="text-embedding-3-small")
```

```python
 from langchain_qdrant import QdrantVectorStore
from qdrant_client import QdrantClient
from qdrant_client.http.models import VectorParams, Distance

client = QdrantClient(path="vectorstore/")

client.create_collection(
    collection_name="warehouse_children",
    vectors_config=VectorParams(size=1536, distance=Distance.COSINE),
)
```

```python
from qdrant_client import models

client.create_payload_index(
    collection_name="warehouse_children",
    field_name="parent_id",
    field_schema=models.PayloadSchemaType.KEYWORD,
)

client.create_payload_index(
    collection_name="warehouse_children",
    field_name="doc_id",
    field_schema=models.PayloadSchemaType.KEYWORD,
)

client.create_payload_index(
    collection_name="warehouse_children",
    field_name="heading",
    field_schema=models.PayloadSchemaType.KEYWORD,
)
```

```python
from langchain_qdrant import QdrantVectorStore

vectorstore = QdrantVectorStore(
    client=client,
    collection_name="warehouse_children",
    embedding=embeddings,
)
```

```python
vectorstore.add_documents(child_docs)
```

```python
def retrieve_parents(query: str, k: int = 4):
    children = vectorstore.similarity_search_with_score(query, k=k)
    seen, parents = set(), []
    score = 0
    for c,score in children:
        pid = c.metadata.get("parent_id")
        if pid and pid not in seen and pid in parent_dict:
            seen.add(pid)
            parents.append(parent_dict[pid])
            score += score
    return {"parents": parents, "score": score}
```

```python
data = retrieve_parents("how to handle Product Damage")
data
```
